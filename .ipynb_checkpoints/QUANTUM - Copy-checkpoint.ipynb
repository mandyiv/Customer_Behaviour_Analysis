```python
import re

# Define regex pattern to extract numbers at the end of each product name
pattern = r'(\d+)[gG]' # Matches one or more digits followed by 'g' at the end of the string

# Extract numbers at the end of each product name using regex and create a new column
df['PROD_WEIGHT'] = df['PROD_NAME'].apply(lambda x: re.search(pattern, x).group(1) if re.search(pattern, x) else None)

# Convert the 'Product_Weight' column to numeric type
df['PROD_WEIGHT'] = pd.to_numeric(df['PROD_WEIGHT'])
```

```python
def count_plot(df, column):
    plt.figure(figsize=(8, 5))
    ax = sns.countplot(x=column, data=df)
    plt.title(f'Countplot of {column}')
    plt.xlabel('{column}')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45, ha ='right')  # Rotate x-axis labels for better readability
    
    # Add count labels on top of each bar
    for p in ax.patches:
        ax.annotate(format(p.get_height(), '.0f'), 
                       (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha = 'center', va = 'center', 
                       xytext = (0, 5), 
                       textcoords = 'offset points')
        
    plt.tight_layout()
    plt.show()

    
```

### Customer Segmentation:
1. Define Customer Segments:

```python
# Define metrics of interest
customer_segments = df.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER']).agg(
    total_sales=('TOT_SALES', 'sum'),
    average_quantity=('PROD_QTY', 'mean'),
    unique_customers=('LYLTY_CARD_NBR', 'nunique')
).reset_index()

customer_segments
```

**Customer Segments:**
The data includes various customer segments such as “MIDAGE SINGLES/COUPLES,” “NEW FAMILIES,” “OLDER FAMILIES,” “OLDER SINGLES/COUPLES,” “RETIREES,” and “YOUNG FAMILIES.”

Each segment has different purchasing behaviors and preferences.

The data is categorized into three premium categories: “Budget,” “Mainstream,” and “Premium.”

**Sales Analysis:**
- “OLDER FAMILIES” in the “Budget” category have the highest total sales ($168,363.25), 
- followed by young singles/couples in the “Mainstream” category ($157621.60).
- “NEW FAMILIES” in the “Premium” category have the lowest total sales ($11491.10).

**Average Quantity Analysis:**
- “OLDER FAMILIES” in the “Premium” category have the highest average quantity (1.98), indicating they purchase more items per transaction.
- “YOUNG SINGLES/COUPLES” in the “Budget” category have the lowest average quantity (1.80).

**Unique Customers Analysis:**
- “YOUNG SINGLES/COUPLES” in the “Mainstream” category have the highest number of unique customers (8,088).
- “NEW FAMILIES” in the “Premium” category have the lowest number of unique customers (588).

**Overall Conclusion:**
- “OLDER FAMILIES” are the highest spenders, especially in the “Budget” category.

- “YOUNG SINGLES/COUPLES” in the “Mainstream” category have a large customer base and are the 2nd highest spenders
- “YOUNG SINGLES/COUPLES” in other categories have a relatively low transaction and sales

- “RETIREES” are significant spenders in the “Mainstream” category.

- The 'New Families' in the premium category have the lowest sales and the lowest number of unique customers

- The “Premium” category generally has higher average quantities, suggesting higher-quality products.


Businesses should tailor marketing strategies based on these insights to maximize revenue and customer engagement.

```python
#lets visualize the metrics

plt.figure(figsize=(12, 6))
ax = sns.barplot(x='LIFESTAGE', y='total_sales', hue='PREMIUM_CUSTOMER', data=customer_segments)
plt.title('Total Sales by Customer Segment')
plt.xlabel('Lifestage')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)

for p in ax.patches:
    ax.annotate('{:.1f}%'.format(p.get_height() / customer_segments['total_sales'].sum() * 100), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha='center', va='center', 
                   xytext=(0, 5),
                   fontsize = 8,
                   textcoords='offset points')
     
        
plt.show()
```

```python
# Calculate total sales, average purchase amount, purchase frequency, etc.
df['Total_Sales_Per_Customer'] = df.groupby('TXN_ID')['TOT_SALES'].transform('sum')
average_purchase = df['TOT_SALES'].mean()
purchase_frequency = df['TXN_ID'].nunique()  # Count unique transaction IDs

# Analyze popular brands and pack sizes
popular_brands = df['BRAND_NAME'].mode().iloc[0]  # Assuming single most popular brand
popular_pack_size = df['PROD_WEIGHT'].mode().iloc[0]

print(f"Average purchase amount: ${average_purchase:.2f}")
print(f"Purchase frequency: {purchase_frequency}")
print(f"Popular brands: {popular_brands}")
print(f"Popular pack size: {popular_pack_size}g")
```

```python

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PROD_WEIGHT', y='TOT_SALES', data=df)
plt.title('Relationship between Pack Size and Total Sales')
plt.xlabel('Pack Size')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Group data by relevant factors (e.g., LIFESTAGE, BRAND_NAME, Pack_Size)
grouped_by_lifestage = df.groupby('LIFESTAGE')

# Calculate descriptive statistics for each group
average_spend_by_lifestage = grouped_by_lifestage['TOT_SALES'].mean()
purchase_frequency_by_lifestage = grouped_by_lifestage['NBR_TXN_ID'].count()

# Print or visualize insights
print(average_spend_by_lifestage.to_string())
print(purchase_frequency_by_lifestage.to_string())

# Explore other groupings and visualizations (e.g., scatter plots to see correlations)
sns.scatterplot(
    x = "Pack_Size",
    y = "TOT_SALES",
    data=df
)
plt.show()
```
